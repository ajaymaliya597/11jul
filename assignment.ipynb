```python
#Q1
numbers = [10,20,4,45,99]
max_number=numbers[0]
i=1
while(i<len(numbers)):
  if numbers[i] > max_number:
    max_number = numbers[i]
  i+=1
print("Max Number : ",max_number)
```

```python
#Q3
numbers = [1,2,3,4,5]
l=len(numbers)-1
i = 0
newl=[]
while i < len(numbers):
  newl.append(numbers[l])
  i += 1
  l -= 1

print("The reverse list is:", newl)
```

```python
# Q4

numbers = [88,5,7,99,55,65]
l = len(numbers) - 1
s = 0

while s < l:
    if numbers[s] > numbers[l]:
        numbers[s], numbers[l] = numbers[l], numbers[s]

    l -= 1

    if s == l:
        l = len(numbers) - 1
        s += 1

print("Ascending list:", numbers)
```

```python
# Q5
numbers = [1, 2, 2, 3, 4, 4, 5]
new = []

for i in numbers:
    if i not in new:
        new.append(i)

print(new)
```
